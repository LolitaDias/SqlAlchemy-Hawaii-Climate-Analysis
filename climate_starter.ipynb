# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func,inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
date_data=session.query(Measurement.date).order_by(Measurement.date.desc()).all()
date_data

#Get the latest date
latest_date=session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(f"The latest date in the Database is : {latest_date[0]}")
```

```python
# Calculate the date 1 year ago from the last data point in the database
#date_one_year_ago=datetime.date(2017,8,23) - datetime.timedelta(days=365)
date_one_year_ago=datetime.datetime.strptime(latest_date[0],'%Y-%m-%d')-datetime.timedelta(days=365)
date_one_year_ago = date_one_year_ago.strftime("%Y-%m-%d")
print(f"The Date 1 year ago from the last data point in the database : {date_one_year_ago}")
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(prcp_data, columns=["Date","Precipitation"])
prcp_df.set_index("Date", inplace=True,)
prcp_df.head()
```

```python
# Design a query to show how many stations are available in this dataset?
stations=session.query(Measurement.station).distinct().count()
print(f" Number of stations available in this dataset: {stations}")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
active_stations = session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc()).all()
active_stations
```

```python
#Which station has the highest number of observations?
print(f"Station with the highest number of observations: {active_stations[0][0]}")
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

temp = session.query(func.min(Measurement.tobs), 
                     func.max(Measurement.tobs),
                    func.avg(Measurement.tobs)).filter(Measurement.station == "USC00519281").all()
print(f"Lowest temperature recorded of the most active station {active_stations[0][0]}: {temp[0][0]}")
print(f"Highest temperature recorded of the most active station {active_stations[0][0]}: {temp[0][1]}")
print(f"Average temperature of the most active station {active_stations[0][0]}: {round(temp[0][2], 1)}")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_data = session.query(Measurement.tobs).\
    filter(Measurement.date >= date_one_year_ago).\
    filter(Measurement.station == "USC00519281").\
    order_by(Measurement.date).all()
temp_data
temp_data_df = pd.DataFrame(temp_data, columns=["tobs"])
temp_data_df.head()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_start_date = input("Enter the trip start date in 'YYYY-MM-DD' format: ")
trip_start_date_obj=datetime.datetime.strptime(trip_start_date,'%Y-%m-%d')
prev_year_date=(datetime.datetime.strptime(trip_start_date,'%Y-%m-%d'))-datetime.timedelta(days=365)
prev_year_date_obj=prev_year_date.strftime('%Y-%m-%d')
print(f"Previous year's date : {prev_year_date_obj}")
trip_data=calc_temps(prev_year_date_obj,trip_start_date_obj) 
tmin=trip_data[0][0]
tmax=trip_data[0][1]
tavg=round(trip_data[0][2])
print(f"Minimum Temperature : {tmin}")
print(f"Maximum Temperature : {tmax}")
print(f"Average Temperature : {tavg}")
```

```python
# Convert the result set to pandas DataFrame
trip_data_df = pd.DataFrame(trip_data, columns=["tmin", "tmax", "tavg"])
trip_data_df
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

y_error = tmax-tmin
trip_data_df.plot.bar(x=0,y="tavg", yerr=y_error, title="Trip Avg Temp", color="coral",legend=None,alpha=0.5,figsize=(2,5))
plt.ylim=(0,50)
plt.xticks([])
plt.ylabel("Temp(F)")
plt.xlabel(" ")
plt.title("Trip Avg Temp")
plt.show()
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
rainfall_per_station = session.query(Measurement.station, func.sum(Measurement.prcp)).\
    filter(Measurement.date >= prev_year_date_obj).\
    filter(Measurement.date <= trip_start_date_obj).\
    group_by(Measurement.station).\
    order_by(func.sum(Measurement.prcp).desc()).all()
rainfall_per_station

# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
sel = [Measurement.station, 
           Station.name, 
           Station.latitude, 
           Station.longitude, 
           Station.elevation, 
           Measurement.prcp]

sorted_result=session.query(*sel).\
            filter(Measurement.station == Station.station).\
            filter(Measurement.date >= prev_year_date_obj).\
            filter(Measurement.date <= trip_start_date_obj).\
            group_by(Measurement.station).\
            order_by(Measurement.prcp.desc()).all()

for res in sorted_result:
    print(res)
```
